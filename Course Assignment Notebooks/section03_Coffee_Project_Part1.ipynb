```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = "coffee/total-production.csv"
coffee_data = pd.read_csv("../Data/coffee/total-production.csv")
new_data = coffee_data.groupby("total_production")
new_data1 = pd.pivot_table(coffee_data, columns=["total_production"])
dis_data = new_data1.sum().sort_values(ascending=False)
news_data = dis_data.astype("int64")
new11_data = news_data.head(10)
figbar = plt.figure(figsize=(10,10))
axbar = figbar.add_subplot()
axbar.bar(new11_data.index, new11_data)
labels = ["BR", "VN", "CO", "IN", "ET", "IND", "ME", "GU", "HO", "UG"]
axbar.set_xticklabels(labels)
axbar.set_ylabel("Production (Millions 60 kg)")
axbar.set_title("Top Coffee Producing Nations 1990 - 2018")
```
